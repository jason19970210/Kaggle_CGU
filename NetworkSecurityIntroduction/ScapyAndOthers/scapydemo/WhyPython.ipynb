Typically, HTML page is composed by tag,  <code>&lt;tag attr="..."&gt;...&lt;/tag&gt;</code>. To extract out *the content* or *attributes* of each **tag**, we use *BeautifulSoup* module for Python3, bs4. 

1. Read the page and analyze with **html**-parser
- find all the possible item in **tr** tag with **class**, like as 
<code style="color:brown;">&lt;tr id=... class="single-book"&gt;</code>


```python
#html = urlopen("http://www.pythonscraping.com/pages/warandpeace.html")
html = urlopen("http://127.0.0.1:8000/book/tw/index.html")
bsObj = BeautifulSoup(html, "html.parser")
nameList = bsObj.findAll("tr", {"class":"single-book"})
for name in nameList:
    print(name.get_text())
```

**First Rule**, the book'info appears at the attribution, <font color="brown">title</font>, within <font color="brown">a</font>-tag, <code style="color:brown;">&lt;a title="book info"&gt;...&lt;/a&gt;</code>; 
; and skip other <font color="brown">a</font>-tag with *null* attribution.

```python
# findAll, get all the items for given condition
nameList = bsObj.findAll("a")
i=1
for name in nameList:
    if name.get('title')==None:
        # pass if no title attribute in <a> 
        1 
    else:
        # print out the book title in order
        print(i,'.',name.get('title'))
        i=i+1
```

**Second Rule**, the book'info also appears *only* at <font color="brown">a</font>-tag followed directly after <font color="brown">h3</font>-tag, like as <code style="color:brown;">&lt;h3&gt;&lt;a title="book info"&gt;...&lt;/a&gt;</code></font>.

```python
# Another short cut, book tilte is also included only in  a-tag after h3-tag
nameList = bsObj.select('h3 a')
i=1
for name in nameList:
    # list all the books in order
    print(i,name.get_text())
    i=i+1
```
