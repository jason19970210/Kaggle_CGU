XSS, Cross-site scripting
---
create your PHP code, named test.php located at  as follows:

```
<?php
   $name = $_GET['name'];
   echo "Welcome $name<br>";
   echo '<a href="http://www.cgu.edu.tw/">Click to Visit CGU</a>';
?>
```

Visit it by your browser:
```
Your_test_site/PHPH/XSS/test.php?name=guest<script>alert('attacked')</script>

```

It would be more dangous if visit by ant malicious codes within itself for instance:
```
Your_test_site/PHPH/XSS/test.php?name=guest<script>location.href="www.google.com";</script>
```

---
