Indent Jump
---
To change the Activity, we have to implement the indent functionality.

1. Acclaim the new indent, MainActivitytwo, in **AndroidManifest.xml**:
```XML
<activity android:name=".MainActivity">
            ...
</activity>
<activity android:name=".MainActivitytwo" android:label="SecondActivity"></activity>
```
where MainActivitytwo(<font color="brown">.kt</font>) is the second activity file created in the following step.
2. Create UI file, activity_main2.xml, and create new button object:
```XML
<?xml version="1.0" encoding="utf-8"?>
<LinearLayout xmlns:android="http://schemas.android.com/apk/res/android"
    ...
    tools:context="com.android.kotlinapp1.MainActivitytwo">
<Button
        android:id="@+id/button4"
        android:layout_width="match_parent"
        android:layout_height="wrap_content"
        android:onClick="onClick_Close"
        android:text="Close"
/>
```
3. Create a new source code, MainActivitytwo.kt:
```java
...
class MainActivitytwo : AppCompatActivity() {

    override fun onCreate(savedInstanceState: Bundle?) {
        super.onCreate(savedInstanceState)
        setContentView(R.layout.activity_main2)
    }
    fun onClick_Close(view: View){
        finish()
    }
}
```
4. To implement the function of jumping from MainActivity to MainActivitytwo, add a new button in activity_main.xml, and create a new function as follows:
```java
import android.content.Intent  
    ...
    fun onClick_showSecond(view: View){
        var intent = Intent(this,MainActivitytwo::class.java)
        startActivity(intent)
    }
```    

How WebView integrated
---
Internet is the one of the best features availed by mobile devices. Not only capable of browsing the web-site contents, but providing the HTML resources by the help of Android UI widget, `webview` and related. 

**OUTLINE**<br>
<pre style="font-family:chalkboard;font-size:1.2em;color:brown;">
AndroidManifest.xml
  acclaim new Activity: WebViewActivity
     â¬‡ï¸Ž
main_activity.xml
  WebView: webview01
     â¬‡ï¸Ž
MainActivity.kt
     â¬‡ï¸Ž
  webview!!.loadUrl(...)  
</pre>
**Details**<br>
1. **acclaim new activity**, <font color="brown">WebViewActivity</font>, in AndroidManifest.xml; and enable <font color="brown">internet</font> permission:
```XML
<manifest xmlns:android="http://schemas.android.com/apk/res/android"
    package="com.android.kotlinapp1">
    <uses-permission android:name="android.permission.INTERNET" />
    ...
    <activity android:name=".MainActivity">
     ...
    </activity>
    <activity android:name=".WebViewActivity" android:label="WebViewActivity"></activity>
    ...
```    
- create <font color="brown">webview.xml</font> in `res/layout`, which contains a webview widget, called `webview01`:
```XML
<?xml version="1.0" encoding="utf-8"?>
<LinearLayout xmlns:android="http://schemas.android.com/apk/res/android"
    xmlns:app="http://schemas.android.com/apk/res-auto"
    xmlns:tools="http://schemas.android.com/tools"
    android:layout_width="match_parent"
    android:layout_height="match_parent"
    android:orientation="vertical"
    tools:context="com.android.kotlinapp1.WebViewActivity"
    >

    <WebView
        android:id="@+id/webview01"
        android:layout_width="fill_parent"
        android:layout_height="fill_parent" />
</LinearLayout>
```
