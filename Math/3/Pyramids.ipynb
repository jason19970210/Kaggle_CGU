To match the physical reality of the reported experiments, [Lubmsn], take the values:
- the observer’s height is $h=1.8$ m, the observers distance $d=10$ m,
- the pyramid’s dimensions $D=23.84$ m, $H=24.02$ m, $q=0.263$ m.
- $\alpha_1=35.01^\circ,  \alpha_2= 78.15^\circ,   \alpha_3  = 82.22^\circ$, and   $\alpha_4 =55.42^\circ$.
- The material properties in the humid Yucatan air have been taken as  $\rho= 1.1466$ kg/m$^3$ for the density and $v=$   343 m/s for the sound velocity.

Those for the limestone14 staircase, [Jacques et.al.], have been taken as $\rho=  2000$ kg/m$^3$ for the density, $v_l=   4100$ m/s for the longitudinal wave velocity and $v_ s =2300$ m/s for the shear wave velocity. Damping has not been taken under consideration.

For the parameters just given, the Lipmann conditions are given as follows: 
- For frequencies lower than 1844 Hz, the numerical simulations will be perfect. 
- For frequencies higher than 1844 Hz, there will be small errors in the description of the sound field within the corrugation, but not elsewhere. 
- For very high frequencies, say more than 5000 Hz, errors may also occur in the prediction of the sound field outside of the corrugation, i.e., in the air and where the observer is situated. The errors gradually grow for higher frequencies and are due to *shadow zones* and neglecting internal reflection within the stairs.

Adopting the data of pyramid dimensions, handclap and sound recording positions by Lubman from Declercq et al., the drawing in above Figure is obtained. The steps of the staircase are numbered $n$=−7 to 84, with $n$=0 being the step at ear  microphone  height. Dimensions are given in meters.

<img src="imgs/stairs-1.png" width=60% />
The sound path length $S_n$  from source-receiver position to step $n$ of the staircase is calculated with Pythagoras’ theorem as follows:

$$
S_n  =   [(11.8 + 0.263n)^2 + (0.263n)^2]^{ 1/2}, n=-7,0,1,2,\cdots,84
$$


Then, with a sound velocity at the site of $v$=343 m/s, the traveling time back and forth, $T_n$   in seconds , follows as
$$
T_n =2\frac{ S_n }{v}, n=−7,\cdots,0,1,2,\cdots,84.
$$

The time interval or delay time  $\tau_n$  between successive reflections at the receiver position then is given by
$$\tau_n=T_{n+1}-T_n,n=−7,\cdots,0,1,2,\cdots,84.$$

Finally, the instantaneous value of the "fundamental" follows as the reciprocal value of  $(n)$ and the “harmonics” fm of the fundamental are given by
$$f_m(n)  = \frac{m}{\tau_n}, m = 1,2,3,4,\cdots$$

```python
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import Audio, Image

%matplotlib inline
```

```python
v=343

nstep=np.arange(-7,85,1)
Sn=np.sqrt((11.8+0.263*nstep)**2+(nstep*0.263)**2)
Tn=2*Sn/v
#taun=np.zeros(len(Tn))
taun=Tn[1:]-Tn[:-1]
```

```python
def FMMod(t):
    return np.cos(2*np.pi*30*t + 5*np.sin(2*np.pi*5*t))
```

```python
# Let's play some sounds.  First play the first reflected echo only 
Fs = 44100 #samplerate
dur = 0.2 #note duration in seconds
t = np.arange(0,dur,1/Fs)

f0=np.sin(2*np.pi*t/taun[7])+np.sin(2*np.pi*t*2/taun[7])+np.sin(2*np.pi*t*3/taun[7])++np.sin(2*np.pi*t*4/taun[7])
Audio(f0, rate=Fs)
```

```python
# play the merged reflected echoes
Fs = 44100 #samplerate
dur = 0.2
#note duration in seconds
t = np.arange(0,dur,1/Fs)

f1=sum(np.sin(2*np.pi*t/tn)+np.sin(2*np.pi*t*2/tn)+np.sin(2*np.pi*t*3/tn)+np.sin(2*np.pi*t*4/tn) for tn in taun)
Audio(f1, rate=Fs)
```

```python
# play the merged reflected echoes in FM modulation
Fs = 44100 #samplerate
dur = 0.2
#note duration in seconds
t = np.arange(0,dur,1/Fs)

f1=sum(np.sin(2*np.pi*t/tn)+np.sin(2*np.pi*t*2/tn)+np.sin(2*np.pi*t*3/tn)+np.sin(2*np.pi*t*4/tn) for tn in taun)
Audio(FMMod(t)*f1, rate=Fs)
```
